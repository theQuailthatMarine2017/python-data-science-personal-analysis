# LINEAR REGRESSION MODEL

__First things first, declare the inputs(Independent Variables/ Predictors) and the targets (dependent variable)__

#### __Below You will notice I have reshaped the data. By default it's a 1D array but scaling and modelling will require it to be 2D. I can do this easily be calling reshape on the values of the inputs and targets. If not an index error will occur__

```python
_input = fifa_clean['Composure'].values.reshape(-1,1)
_target = fifa_clean['Overall'].values.reshape(-1,1)
```

### SCALE DATA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(_input)
```

```python
input_scaled = scaler.transform(_input)
```

### TRAIN TEST SPLIT THE MODEL

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input_scaled, _target,test_size=0.2, random_state=365)
```

### CREATE THE REGRESSION

```python
reg = LinearRegression()

reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat,s=15)
plt.xlabel('Target(y_train)')
plt.ylabel('Predictions(y_hat)')

plt.show()
```

__Everything above looks good. The plot stays close to the 45 degree line that shows linearilty__

### RESIDUALS PDF

__Residuals is the difference between the targets and the predictions.__

__Below I will plot the residuals on a graph using seaborn to see how it is distributed well. just like I did earlier when exploring the features I chose.__

```python
sns.distplot(y_train - y_hat)
plt.title('Training data Residuals PDF',size=10)
```

__Above shows that the Residual is normally distributed with no outliers and a mean of zero.__

__From my knowledge this is a okay sign. Next I will get the R Squared which will help tell us how much of our model is explaining the variabilty of the data__

### CHECK R-SQUARED

```python
reg.score(x_train,y_train)
```

### weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

## ADJUSENT R SQUARDED TO SEE IF WE HAVE BEEN PENALISED FOR USING THE CHOSEN INPUT

```python
r2 = reg.score(x_train,y_train)
n = x_train.shape[0]
p = x_train.shape[1]

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

## TEST

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,s=5,alpha=0.2)
plt.xlabel('Target(y_test)')
plt.ylabel('Predictions(y_hat_test)')

plt.show()
```

## Plottting the Residuals for Test Data. Good sign if it matches the Residual for the training data

```python
sns.distplot(y_test - y_hat_test)
plt.title('Training data Residuals PDF',size=10)
```

### Now I Will Create A DataFrame with The Test Values against the Predictions from the Test Values to See how close the predictions are.

```python
df_pf = pd.DataFrame(y_hat_test,columns=['Overall Predictions'])

df_pf.head()
```

```python
df_pf['Overall Targets'] = y_test

df_pf.describe()
```

__The above description clearly shows that the predicted values we have from our model are close with the test data we provided meaning that the training data did good__


__NB: THIS IS A SIMPLE LINEAR REGRESSION MODEL AND AFTER THIS WILL USE THE MULTIPLE LINEAR REGRESSION TO CREATE ANOTHER MODEL USING DIFFERENT HYPOTHESIS__

__Adding a new columns,RESIDUAL & DIFFERENCES% to the new dataframe which will contain the residual (Difference between the Target Values and the Predictions Obtained) which will help see if the SSE is low which is needed__

```python
df_pf['Residuals'] = df_pf['Overall Predictions'] - df_pf['Overall Targets']
df_pf['Difference in %'] = np.absolute(df_pf['Residuals']/df_pf['Overall Targets']%100)

df_pf.head()
```
